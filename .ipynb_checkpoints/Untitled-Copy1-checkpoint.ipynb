```python
import pandas as pd

df = pd.read_csv("housing_data.csv") 
features_df=df.columns
df.head()
```

```python
print(features_df)
```

```python
# As Suburb is written after , in address. Let's extract and make a different column
df['Suburb'] = df['Address'].apply(lambda x: x.split(',')[-1].strip() if pd.notnull(x) else None)

print(df.head())
```

```python
import pandas as pd
import re
import numpy as np
# Let's clean Price value as their is range given for some values. So, we will take average of those values and also remove symbols to make it numeric.
def clean_price(value):
    if pd.isnull(value):
        return np.nan
    
    text = str(value)
    
    # Handling price ranges by taking the average
    if "-" in text:
        nums = re.findall(r'\d[\d,]*', text)
        if len(nums) == 2:
            low = int(nums[0].replace(",", ""))
            high = int(nums[1].replace(",", ""))
            return (low + high) / 2

    # Handling single price by removing the symbols like , and $
    nums = re.findall(r'\d[\d,]*', text)
    if nums:
        return int(nums[0].replace(",", ""))

    return np.nan

# clean each price by applying clean_price() function to each price value
df['Price'] = df['Price'].apply(clean_price)

print(df['Price'].head(10))
```

```python
df = df.dropna(subset=['Address', 'Price'])
print(df)
```

```python
df['Area'] = df['Area'].str.extract(r'(\d+\.?\d*)').astype(float)
num_cols = ['Bedrooms','Bathrooms','Carspaces','Price','Area']
# Now, convert the each NaN value to median of the column because median is not affected by outliers.
for col in num_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

for col in num_cols:
    df[col] = df[col].fillna(df[col].median())
# Since, address and name are compuslsory and we already removed the rows which don't have them
# We will handle NaN value of Type column now;
df['Type'] = df['Type'].fillna("Unknown")
```

```python
# firstly, fill the Nan value but this will change column type to str not date
df['Date_time'] = df['Date_time'].fillna("Inspection Thu 18 Sep 12:00 pm")
# removing the inspection 
df['demo'] = df['Date_time'].str.replace(r'^Inspection\s+', '', regex=True)

df['Date_time_parsed'] = pd.to_datetime(
    df['demo'], format='%a %d %b %I:%M %p', errors='coerce'
)

default_dt = pd.to_datetime('2025-09-18 12:00 pm')
df['Date_time_parsed'] = df['Date_time_parsed'].fillna(default_dt)

# Now, extract the date time seperately too
df['Date'] = df['Date_time_parsed'].dt.date
df['Time'] = df['Date_time_parsed'].dt.time

# dropping unnecessary columns
df = df.drop(columns=['demo'])
df = df.drop(columns=['Date_time'])
```

**2. Conduct data preprocessing and exploratory data analysis:**
**a. Convert categorical variables (e.g. unit/house/apartment) using one-hot encoding.**

Now, we will apply one-hot encoding to Type and Suburb column of the dataset. Since, address and Name are unique for each and they are not the reason the price will change. So, we are not using these columns in the model. 

```python

cat_cols = ['Type', 'Suburb']  
df= pd.get_dummies(df, columns=cat_cols, drop_first=True)
```

**b. Create new features (e.g., number of schools nearby).**

```python
# Let's create new features
# One feature Suburb was already created in the first part from address column
df['Total_rooms']=df['Bedrooms']+df['Carspaces']
df['Price_of_Bedroom'] = df['Price'] / (df['Total_rooms'] + 1)
```

```python
df.head(10)
```

```python
# Now, add nearby schools too but since, in real dataset nobody mentioned. we can generate value using random numbers functios of numpy.
import numpy as np
np.random.seed(100)
df['Schools'] = np.random.randint(0, 10, size=len(df))
```

**c. Normalize or standardize numerical features.**

Before modelling let's delete address and name column as they are not needed in modelling.

```python
df = df.drop(columns=['Address', 'Name'])
df.head(10)
```

```python
num_scaled=['Bedrooms','Bathrooms','Carspaces','Area','Total_rooms','Price_of_Bedroom','Schools']
# Let's standardize each numeric column so each column have same mean and variance
# We didn't scaled price as it was response variable
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df[num_scaled] = scaler.fit_transform(df[num_scaled])
import pickle
with open("scaler.pkl", "wb") as f:
    pickle.dump(scaler, f)
df.head(10)
```

**e. Identify price trends over time across suburbs.**

```python
suburb_cols = [c for c in df.columns if c.startswith('Suburb_')]

# Get the suburb name
df['Suburb'] = df[suburb_cols].idxmax(axis=1).str.replace('Suburb_', '')

# Create month column
df['Month'] = df['Date_time_parsed'].dt.to_period('M')

trend = df.groupby(['Suburb', 'Month'])['Price'].mean().reset_index()
trend['Month'] = trend['Month'].dt.to_timestamp()  

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,6))
sns.lineplot(data=trend, x='Month', y='Price', hue='Suburb')
plt.title("Trend of Price over Time accross Suburb")
plt.ylabel("Price")
plt.xlabel("Month")
plt.xticks(rotation=45)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

Since, our dataset has included nearby regions too that's why it has a lot of suburbs.

**3. Model development: develop at least three different regression models, use MAE, RMSE and R-squared as the evaluation metrics, use k-fold cross validation to evaluate the model performance.**

```python
from sklearn.model_selection import train_test_split
# predictor variable
# we will exclude suburb as we added to plot and now we already have one hot columns for that
df = df.drop(columns=['Suburb']) 
# since we can't model on date time we will drop that too

X = df.drop(columns=['Price','Month','Time','Date','Date_time_parsed']) 
#response variable
y = df['Price']                 

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=100)
X.head()
```

```python
#Let's develop linear regression, Random forest and gradient booster and then evaluate them
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.model_selection import cross_val_score, KFold
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np
#let's fit the models
models = {
    "Linear Regression": LinearRegression(),
    "Random Forest": RandomForestRegressor(random_state=100),
    "Gradient Boosting": GradientBoostingRegressor(random_state=100)
}
# Now, evaluate the model performance
# We will use cross-validation and split data into 5 folds
folds = KFold(n_splits=5, shuffle=True, random_state=100)
# shuffling is needed if we want to set random_state"
for name, model in models.items():
    # cross_val_score runs the model 5 times
    # Since, sklearn gives negative error values so negative
    mae = -cross_val_score(model, X, y, cv=folds, scoring='neg_mean_absolute_error').mean()
    # rmse is sqrt of mean square error
    rmse = np.sqrt(-cross_val_score(model, X, y, cv=folds, scoring='neg_mean_squared_error').mean())
    r2_val = cross_val_score(model, X, y, cv=folds, scoring='r2').mean()
    
    print(f"{name}: MAE={mae}, RMSE={rmse}, R2={r2_val}\n")
```

We can interpret that Mean absolute error of random forest is least indicating predictions are close to actual values. Root mean square error of gradient boosting is minimum meaning it is better than other models. However, the R^2 value is almost same for all the models i.e. all the models explain approximately 99% of the variance. Hence, if we want to decrease large errors and reduce outliers choose gradient boosting otherwise random forest performs well to decrease average error values.

**4. Feature importance: identify which features more influence housing price, use model-specific methods (e.g., feature importance in tree-based methods) or other statistical methods (e.g., SHAP values).**

Since, both Random Forest and Gradient Boosting are tree based models we will use model specific methods. Here, I am using Random Forest as I want to decrease average error values and not only large errors.

```python
import matplotlib.pyplot as plt
# Fitting random forest model
RF_model = RandomForestRegressor(random_state=100)
RF_model.fit(X, y)

# Now, check the importance of each feature
importances = RF_model.feature_importances_
features = X.columns

# Let's Sort the according to importance
index= np.argsort(importances)[::-1] # -1 to sort in descending order

for i in index:
    print(f"{features[i]}: {importances[i]}")
```

```python
# Since, we know the features that influence more let's make new model and deploy that so, that the features are lesser as the previous 
X=df[['Area','Bedrooms','Bathrooms','Schools']]
y=df['Price']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=100)

model_ = RandomForestRegressor(random_state=100)
model_.fit(X_train, y_train)

# Save the model now,
with open("RF_model_.pkl", "wb") as f:
    pickle.dump(model_, f)
```

**5.
Model deployment: Develop a simple web demo application using appropriate packages (e.g., gradio, Flask or Streamlit). Allow users to input property features and receive price predictions.**


```python
# Save your Streamlit app code to a .py file from the notebook
code = """
import streamlit as st
import pandas as pd
import pickle

# Loading training model
model = pickle.load(open('RF_model_.pkl', 'rb'))

st.title("Housing Price Prediction")

# User inputs
Area = st.number_input('Area (sqft)')
Bedrooms = st.number_input('Number of Bedrooms')
Bathrooms = st.number_input('Number of Bathrooms')
Schools = st.number_input('Number of Schools Nearby')


input_data = pd.DataFrame([[Area, Bedrooms, Bathrooms, Schools]],
                          columns=['Area', 'Bedrooms', 'Bathrooms', 'Schools'])
scaler = pickle.load(open("scaler.pkl", "rb"))

input_data[['Area', 'Bedrooms', 'Bathrooms', 'Schools']] = scaler.transform(input_data)

# Prediction
if st.button('Predict Price'):
    prediction = model.predict(input_data)
    st.success(f"Predicted Price: ${prediction[0]:,.2f}")
"""

with open("app.py", "w") as f:
    f.write(code)
```
